```python
#Embedding with custom dimension
embeddings_1024=OpenAIEmbeddings(model="text-embedding-3-large", dimensions=1024)
```

```python
embeddings_1024
```

```python
## Text Loder
from langchain_community.document_loaders import TextLoader

loader = TextLoader(r'E:\GenAI_Krish\Sec23_1-Basics+Of+Langchain\3.2-DataIngestion\speech.txt')
docs=loader.load()
docs
```

```python
### Recursively split text by characters
from langchain_text_splitters import RecursiveCharacterTextSplitter

text_splitter= RecursiveCharacterTextSplitter(chunk_size= 500, chunk_overlap=50)
final_documnet=text_splitter.split_documents(docs)
final_documnet
```

```python
## Vector Embedding And Vector StoreDB
from langchain_community.vectorstores import Chroma

db= Chroma.from_documents(final_documnet, embeddings_1024)
db
```

```python
query = "It will be all the easier for us to conduct ourselves as belligerents"
retrive_result= db.similarity_search(query)
print(retrive_result)
```
